```python
#load the data
import pandas as pd

df = pd.read_csv('sample.csv')

print(df.to_string())
```

```python
#Tokenization
import nltk
nltk.download('punkt')
```

```python
#Tokenization
from nltk import word_tokenize
text = df.to_string()
words = word_tokenize(text)
print(words)
```

```python
#lower casing
text_lower = []
for t in words:
    text_lower.append(t.lower())
print(text_lower)

```

```python
#Punctuation Removal
import string
text_p = [char for char in text_lower if char not in string.punctuation]
print(text_p)
```

```python
#download stopword library
import nltk
nltk.download('stopwords')
```

```python
#Removal of stop words
from nltk.corpus import stopwords
stop_words = stopwords.words('english')

filtered_words = [word for word in text_p if word not in stop_words]
print(filtered_words)
```

```python
#Stemming
from nltk.stem.porter import PorterStemmer
porter = PorterStemmer()
stemmed = [porter.stem(word) for word in filtered_words]
print(stemmed)
```

```python
#download Lemmatization library
import nltk
nltk.download('wordnet')
```

```python
#download Lemmatization library
import nltk
nltk.download('omw-1.4')
```

```python
#Lemmatization
from nltk.stem import WordNetLemmatizer
  
lemmatizer = WordNetLemmatizer()
  
lemmatized = [lemmatizer.lemmatize(word) for word in filtered_words]
print(lemmatized)
```

```python
#join text
re_text = " ".join(lemmatized)
print(re_text)
```

```python
##Removal of emojis, emoticons, and URLs

import re

emoji_pattern = re.compile("["
        u"\U0001F600-\U0001F92F"  # emoticons
        
                           "]+", flags=re.UNICODE)

no_emoji = emoji_pattern.sub(r'', re_text)

no_html = re.sub(r'http', r'', no_emoji)

result = re.sub(r'/\S+', '', no_html)
print(result) # no url
```
